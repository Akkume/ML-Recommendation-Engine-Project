```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python


userInput = [
            {'title':'Snake', 'rating':3.5},
            {'title':'Kesariya', 'rating':3},
            {'title':'Starboy', 'rating':5},
         ]
inputsongs=pd.DataFrame(userInput)
inputsongs
```

```python
songs_genres = ['Pop','Jazz','Electronic','Classical']
songs = [['Snake', 1,1, 0, 0], ['Kesariya', 0,1, 1, 0],['Starboy', 0, 1, 0,1]]
songs_df = pd.DataFrame(songs, columns = ['Title'] + songs_genres)
songs_df
```

```python
C = songs_df[['Pop','Jazz','Electronic','Classical']].to_numpy()
C
```

```python
u=np.array([3.5,3,5])
u
```

```python
unkown=[['Ranja',0,1,0,1],['Mainke',0,0,1,1],['Excuses',1,1,0,0]]
new_songs_df = pd.DataFrame(unkown, columns = ['Title']+songs_genres)
new_songs_df
```

```python
D=np.matmul(u,C)
```

```python
sum=0
for i in userInput:
  for j in i:
    if j=='rating':
      sum=sum+i[j]

print(sum)
```

```python
t=[]
for i in D:
  i=i/sum
  t.append(i)
np.array(t)
```
